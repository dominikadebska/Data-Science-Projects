Limiting dataset to a period from 1850 to 2020

```python
data_1850 = data[data["year"] >= 1850]
```

```python
data_median = data_1850.groupby("year").median().reset_index()
```

### CO2 comparison for selected countries

```python
data_Poland = data_1850[data_1850["country"] == "Poland"].groupby("year").median().reset_index()
```

```python
data_Norway = data_1850[data_1850["country"] == "Norway"].groupby("year").median().reset_index()
```

```python
data_Brazil = data_1850[data_1850["country"] == "Brazil"].groupby("year").median().reset_index()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(data_median["year"], data_median["co2_per_capita"])
plt.plot(data_Poland["year"], data_Poland["co2_per_capita"])
plt.plot(data_Norway["year"], data_Norway["co2_per_capita"])
plt.plot(data_Brazil["year"], data_Brazil["co2_per_capita"])
plt.xlabel("Year")
plt.ylabel("CO2 consumption per capita")
plt.legend(["Total", "Poland", "Norway", "Brazil"], loc = 1)
plt.xlim(1845, 2020)
plt.title("CO2 per capita [Median]")
```

We see that over the last few decades, co2 emission per capita has doubled â€” from approximately 2 (metric tones) in 1980, to around 4 in 2020. When we dive deeper on a country level, we can see that e.g. in Poland the peak was reached in 1980 as the co2 per capita was amounted to 12 but in the next few decades, the numbers were going down and leveled off at point 8.
